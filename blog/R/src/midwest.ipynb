```r
midwest1 <- midwest
midwest1 <- rename(midwest1, total=poptotal)
midwest1 <- rename(midwest1, asian=popasian)
names(midwest1)
```

```r
midwest1$total
```

```r
midwest1$asian
```

##### poptotal(전체인구) 변수를 total로 popasin(아시나 인구)변수를 asian으로 수정

##### toal, asian 변수를 이용하여 "전체인구 대비 아시아 인구 백분율" 파생변수를 만들고, 히스토그램을 만들어 분포를 살펴본다

```r
a <- (midwest1$asian/midwest1$total) *100
a <- round(a,2)
hist(a)
```

```r
names(midwest1)
```

```r
mean(a)
```

##### 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 "large", 그 외는 "small"을 부여하는 파생변수를 만드시오

```r
midwest1$ratio <- a
b <- mean(a)
midwest1$group <- ifelse(midwest1$ratio > b, "large", "small")
```

```r
head(midwest1,2)
```

##### "large"와 "small"에 해당하는 지역이 얼마나 되는지 빈도표와 빈도 막대그래프를 만드시오

```r
table(midwest1$group)
```

```r
qplot(midwest1$group)
```
