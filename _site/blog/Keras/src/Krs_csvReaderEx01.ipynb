```python
from keras.models import Sequential
from keras.layers.core import Dense
from keras import optimizers

import numpy as np
```

```python
data = np.loadtxt('../data/score2.csv',delimiter=',', dtype=np.float32)
```

```python
table_row = data.shape[0]
table_col = data.shape[1]

y_column = 1
x_column = table_col - y_column # 입력 데이터의 컬럼 개수

# 모든 행의 앞 3개는 입력으로 본다.
x_data = data[:,0:x_column]
y_data = data[:,x_column:(x_column+1)]

x_test = [[20,40,50],[90,88,80]]
```

```python
model = Sequential()

model.add(Dense(units=1, input_dim=x_column))
model.compile(loss='mse', optimizer='rmsprop')
model.fit(x_data, y_data, epochs=2000)
```

```python
for item in x_test:
    result = model.predict(np.array([item]))
    print('-'*50)
    print(result)
```
