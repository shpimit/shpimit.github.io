# 뉴스 사이트(https://news.naver.com/)에 requests로 가져오기
* 1번 사이트에서 "가장 많이 본 뉴스" 하위 6개 카테고리 찾기
* 각 카테고리 별 10개의 뉴스 링크 확인
* 뉴스링크 주소 정규화
* 각 뉴스 다운로드
* 뉴스 본문 영역 추출
* 텍스트만 저장
* 파일이름 : 카테고리-고유번호(aid).txt
* 총 6개의 카테고리, 60개를 수집

```python
import requests
```

```python
def getDownload(url, param = None, retries = 3):
    resp = None
    
    try:
        resp = requests.get(url, params = param, headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36'})
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e:
        if 500 <= resp.status_code < 600 and retries > 0:
            print('Retries : {0}'.format(retries))
            return getDownload(url, param, retries -1)
        else:
            print(resp.status_code)
            print(resp.reason)
            print(resp.request.headers)
            
    return resp
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://news.naver.com/"
html = getDownload(url)
dom = BeautifulSoup(html.text,"html.parser")

from urllib import parse
import sys
```

```python
for tag in dom.select("div.main_aside .section_list_ranking a"):
    print("*"*100)
##    print(tag)
    print(tag.find_parent().find_parent().find_parent().text[0])
    tag1 = tag.find_parent().find_parent().find_parent().h5.text
    print(tag1)
##    print(tag.text)
    if(tag.has_attr("href")):
        newUrl = requests.compat.urljoin(url,tag["href"])
        print(newUrl)
        pnewUrl = parse.urlparse(newUrl).query
##        print(pnewUrl)
##        print('>'+pnewUrl[53:63])
        newsHtml = getDownload(newUrl)
        newsdom = BeautifulSoup(newsHtml.text,"html.parser")
        print("#"*100)
##        print(newsdom.select("#articleBodyContents"))
        for newstag in newsdom.select("#articleBodyContents"):
##            print(newstag.text)
            filename = tag1+"-"+pnewUrl[53:63]+".txt"
            newfilename = filename.replace("/","")
            print(newfilename)
            file = open("../data/"+newfilename, 'w', encoding='utf8')
            file.write(newstag.text)
            file.close()
##    break 
```
