```python
frame = DataFrame(np.arange(9).reshape(3,3),
                  index = ['강감찬','김유신','광해군'],
                  columns=['서울','부산','광주'])
frame
```

```python
# 인덱스 재조정
frame = frame.reindex(['강감찬','이순신','김유신','광해군'])
frame
```

```python
states = ['서울','부산','울산']
frame = frame.reindex(columns = states)
frame
```

```python
# ix를 이용하게 되면 라벨로 색인하면 좀 더 간결하게 할 수 있다.
# 특정한 열,행만 보고싶을때 사용함.
frame = frame.ix[['강감찬','이순신','김유신','광해군'], states]
frame
```

```python
# ix 대신에  loc 사용
states = ['서울','부산']
myindex = ['강감찬','김유신','광해군']
frame = frame.loc[myindex, states]
frame
```

```python
states = ['서울','부산']
myindex = ['강감찬','김유신','광해군']
frame = frame.loc[myindex, states]
frame
```

```python
myframe = DataFrame( (np.arange(12).reshape(4,3)),
                     index=[
                         ['주니어','주니어','테일러','테일러'],
                         ['kor','eng','kor','eng']
                         ],
                     columns = [
                         ['서울','부산','서울'],
                         ['Center','Out','Center']])

myframe.index.names = ['names','subject']
myframe.columns.names = ['city','region']
```

```python
myframe = DataFrame(np.arange(16).reshape(4,4),
                    index = [['주니어','주니어','테일러','테일러'],['사과','수박','사과','수박']],
                    columns=[['서울','부산','서울','부산'],['Green','Green','Red','Red']])

myframe
```

```python
myframe.index.names = ['name', 'fruit']
myframe.columns.names = ['city', 'color']
myframe
```

```python
newframe = myframe.sum(level='name')
newframe
```

```python
newframe = myframe.max(level='fruit')
newframe
```

```python
# 도시별 평균 값
newframe = myframe.mean(level='city', axis=1)
newframe
```

```python
# 색상별 최소 값
newframe = myframe.min(level='color', axis=1)
newframe
```

```python
newframe = myframe.sum(level='name').mean(level='city',axis=1)
newframe
```
